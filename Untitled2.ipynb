```python
class Account:
    def __init__(self, user_id, pin):
        self.user_id = user_id
        self.pin = pin
        self.balance = 0
        self.transaction_history = []

    def deposit(self, amount):
        self.balance += amount
        self.transaction_history.append(f'Deposit: +${amount}')

    def withdraw(self, amount):
        if self.balance >= amount:
            self.balance -= amount
            self.transaction_history.append(f'Withdrawal: -${amount}')
        else:
            print("Insufficient funds.")

    def transfer(self, recipient, amount):
        if self.balance >= amount:
            self.balance -= amount
            recipient.balance += amount
            self.transaction_history.append(f'Transfer to {recipient.user_id}: -${amount}')
            recipient.transaction_history.append(f'Transfer from {self.user_id}: +${amount}')
        else:
            print("Insufficient funds.")

    def display_transaction_history(self):
        for transaction in self.transaction_history:
            print(transaction)

def main():
    user_id = input("Enter User ID: ")
    pin = input("Enter PIN: ")

    # Create an account
    account = Account(user_id, pin)

    while True:
        print("\nATM Options:")
        print("1. View Transaction History")
        print("2. Deposit")
        print("3. Withdraw")
        print("4. Transfer")
        print("5. Quit")

        choice = input("Enter your choice: ")

        if choice == '1':
            account.display_transaction_history()
        elif choice == '2':
            amount = float(input("Enter the deposit amount: $"))
            account.deposit(amount)
        elif choice == '3':
            amount = float(input("Enter the withdrawal amount: $"))
            account.withdraw(amount)
        elif choice == '4':
            recipient_id = input("Enter recipient's User ID: ")
            recipient = Account(recipient_id, None)  # For simplicity, we omit PIN verification here
            amount = float(input("Enter transfer amount: $"))
            account.transfer(recipient, amount)
        elif choice == '5':
            print("Goodbye!")
            break
        else:
            print("Invalid choice. Please try again.")

if __name__ == "__main__":
    main()
```
